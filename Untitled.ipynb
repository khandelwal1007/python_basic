```python
# 6. How do conditional statement work in python? illustrate with examples.

Conditional statement: Conditional statements in Python allow you to execute certain blocks of code based on specific conditions. They are fundamental in controlling the flow of your program.
1.if
2.if-else
3.if-elif-else
4.nested-if-else

1.if statement: if statement check a condition and execute a block of code if the condition is 'True'.
Example: grade = 8
         co_cirr = True
         if ((grade >= 8) and (co_cirr == True)):
            print("you are all rounder")
            
2. if-else statement: if-else statement provides an alternative block of code to execute when the condition is False.
example:
temperature = 18

if temperature > 25:
    print("It's a hot day!")
else:
    print("It's not a hot day.")
    
3. if-elif-else statement: if-elif-else statement allows you to check multiple conditions in sequence.
Example : 
temperature = 20

if temperature > 30:
    print("It's a very hot day!")
elif temperature > 25:
    print("It's a hot day!")
elif temperature > 15:
    print("It's a mild day!")
else:
    print("It's a cold day.")


    
4.nested-if-else statement: You can also nest conditional statements inside one another for more complex logic.
example:
temperature = 22
weather = "sunny"

if temperature > 20:
    if weather == "sunny":
        print("It's a warm and sunny day!")
    else:
        print("It's warm, but not sunny.")
else:
    print("It's not a warm day.")
```
